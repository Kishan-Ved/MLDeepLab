```python
# K Center Clustering

# Output - Cluster centers, plot of objective value vs no. of iterations 
import numpy as np
from sklearn.datasets import load_digits  
import matplotlib.pyplot as plt

digits = load_digits().data
n,m = digits.shape
r = np.random.randint(0,n)
k = 10
center = [r]
label = [] # Centers of each element in digits
obj = [] # Objective values
distances = [] # Individual distances
for _ in range(k-1):
  d = 0 
  label = []
  distances = []

  # Assigning the center
  for i in range(n):
    m1 = np.inf
    c1 = 0
    for j in range(len(center)):
      d = np.linalg.norm(digits[i]-digits[center[j]])
      if d<m1:
        m1 = d
        c1 = j
    label.append(center[c1])

    distances.append(np.linalg.norm(digits[i]-digits[label[i]]))
  ind = np.argmax(distances)
  center.append(ind)
  obj.append(distances[ind])
print(center)
print(obj[-1])
plt.plot(np.arange(2,11),obj)
plt.xlabel("No. of Centers")
plt.ylabel("K Center Objective")
plt.title("K Center Clustering")
plt.show()
fig, ax = plt.subplots(2,5,figsize=(8,3))
cent = []
for i in center:
  cent.append(digits[i])
cent = np.array(cent)
cent = cent.reshape(10,8,8)
for axi,cen in zip(ax.flat,cent):
  axi.set(xticks=[], yticks=[])
  axi.imshow(cen,interpolation='nearest',cmap=plt.cm.binary)
```
